```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error
x = dt.drop("median_house_value", axis=1)
y = dt["median_house_value"]
x = pd.get_dummies(x)


x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
# prepruning
dt_pre = DecisionTreeRegressor(
    max_depth=5,
    min_samples_split=10,
    min_samples_leaf=5,
    random_state=42
)
dt_pre.fit(x_train, y_train)
y_pred_pre = dt_pre.predict(x_test)
mse_pre = mean_squared_error(y_test, y_pred_pre)
print("Pre-Pruned MSE:", mse_pre)
# postpruning
# Train full tree
dt_full = DecisionTreeRegressor(random_state=42)
dt_full.fit(x_train, y_train)

path = dt_full.cost_complexity_pruning_path(x_train, y_train)
ccp_alphas = path.ccp_alphas

# train multiple pruning trees
mse_list = []

for alpha in ccp_alphas:
    dt = DecisionTreeRegressor(random_state=42, ccp_alpha=alpha)
    dt.fit(x_train, y_train)
    y_pred = dt.predict(x_test)
    mse_list.append(mean_squared_error(y_test, y_pred))

# Find the alpha that gives the lowest MSE
best_alpha = ccp_alphas[np.argmin(mse_list)]
print("Best alpha:", best_alpha)

dt_post = DecisionTreeRegressor(random_state=42, ccp_alpha=best_alpha)
dt_post.fit(x_train, y_train)

y_pred_post = dt_post.predict(x_test)
mse_post = mean_squared_error(y_test, y_pred_post)
print("Post-Pruned MSE:", mse_post)
```
